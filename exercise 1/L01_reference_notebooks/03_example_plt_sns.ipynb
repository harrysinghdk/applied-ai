```python
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

%%

# quick theming for dark backgrounds plots

```python
sns.set_style("darkgrid")
matplotlib.rcParams["axes.edgecolor"]   = "white"
matplotlib.rcParams["axes.labelcolor"]  = "white"
matplotlib.rcParams["axes.titlecolor"]  = "white"
matplotlib.rcParams["grid.color"]       = "white"
matplotlib.rcParams["xtick.color"]      = "white"
matplotlib.rcParams["xtick.labelcolor"] = "white"
matplotlib.rcParams["ytick.color"]      = "white"
matplotlib.rcParams["ytick.labelcolor"] = "white"
matplotlib.rcParams["figure.facecolor"] = "1f1f1f"
matplotlib.rcParams["savefig.transparent"] = True
```

%%

```python
df = pd.read_csv("data/california_housing_train.csv").head(1000)
```

%%<br>
makes a lineplot (and renders it, if in notebook)

```python
sns.lineplot(df, x="latitude", y="longitude")
```

saves the plot to file<br>
IMPORTANT: in notebooks, savefig must appear in the same cell as the plot,<br>
otherwise it won't work

```python
plt.savefig("test_plot.png")  
                        
```

```python
sns.lineplot(df, x="latitude", y="latitude")
sns.lineplot(df, x="latitude", y="longitude")
```
