```python
c = np.random.random(size=(10, 2))  # uniform random
```

```python
c = c * 2 - 1   # operators are overloaded appropriately
                # what does this operation do?
```

```python
d = c * np.random.normal(       # IMPORTANT! we are wirking on multidimensional vectors, NOT matrices!
    size=(10, 2),               # `*` implementes element-wise multiplication, same for most other operations
    loc=1,                      # in this case, we are multiplying with a random normal distribution with
    scale=.2                    # mean=1 and std=.2
) 
# c = c * b this would raise an error, sice c.shape == (10, 2) but b.shape == (10, 3)
```

if you actually need a matrix multiplication

```python
m = np.matmul(c, np.random.random(size=(2, 10)))
```
