## 1. Plotting simple functions

Example: plotting `sin(x ** 2)` for `x` between `-pi` and `pi`:

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-np.pi, np.pi, 100)
y = np.sin(x ** 2)

plt.plot(x, y)
```
